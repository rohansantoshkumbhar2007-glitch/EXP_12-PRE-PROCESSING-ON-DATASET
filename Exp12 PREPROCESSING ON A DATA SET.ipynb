```python
#Dataset-1
import pandas as pd
data= { "Roll No":[101,102,103,104,105],
       "Name":["Ayush","Aaryan","Aditya","Harsh","Anish"],
        "Age":[20,"-",21,20,"22"],
        "Marks":[85,90,"-",88,76],
        "Department":["CSE","IT","CSE","CSE","E&TC",],
        "Admission_Date":["01-06-2023","02-06-2023","04-06-2023","06-06-2023","08/06/2023"]
        }
df=pd.DataFrame(data)

#Save as CSV
df.to_csv("student_data.csv",index=False)



```

```python
#Step 2. Replace '-' with NaN to mark missing values

df.replace("-",np.nan,inplace=True)

#Convert colmns to numeric where needed

df["Age"]=pd.to_numeric(df["Age"],errors='coerce')
df["Marks"]=pd.to_numeric(df["Marks"],errors='coerce')
df

### When converting data(Like strings--> datetime),
### if some valuues are invalid or cannot be converted,
### instead of thowing an error, it wll:
### ➡️Replace those invalid values wi NaT (Not a Time)'''
```

```python
# Step 3. Check for missing values

print("\nMissing VaLues: ")
print(df.isnull().sum())
```

```python
#Fill missing age with mean

df["Age"].fillna(df["Age"].mean(),inplace=True)
df

```

```python
#Fill missing marks with median
df["Marks"].fillna(df["Marks"].median(), inplace = True)
df
```

```python
print(df.isnull().sum())
```

```python
#Fix inconsistent text formatting
df["Department"] = df["Department"].str.upper()
df
```

```python
#Step 6 - Convert Admission_Date to standard datetime format

df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors='coerce')
df

```

```python
#Step 8: Save cleaned dataset

df.to_csv("cleaned_student_data.csv", index=False)

print("\nCleaned CSV file saved as cleaned_student_data.csv")
df
```

```python
#Working with missing values on Dataset-2
#Upload the Dataset

import pandas as pd
import numpy as np
df=pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum()  #Attributes are displayed
```

```python
#Imputing the missing value
#1.Luggage Room
df['Luggage.room']=df['Luggage.room'].fillna(value=df['Luggage.room'].mean())
```

```python
#2. Rear seat room
df['Rear.seat.room']=df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
df
```

```python
df['AirBags'] = df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
df.isna().sum()
```

```python
#Save Cleaned dataset
df.to_csv("cleaned_Cars93.csv",index=False)
df
```
